```python
#References
#GeeksforGeeks. (2018). ML | Label Encoding of datasets in Python. [online] Available at: https://www.geeksforgeeks.org/ml-label-encoding-of-datasets-in-python/ [Accessed 28 Feb. 2023].
#Velardo, V. (2020). Deep Learning (for Audio) with Python - YouTube. [online] www.youtube.com. Available at: https://www.youtube.com/playlist?list=PL-wATfeyAMNrtbkCNsLcpoAyBBRJZVlnf [Accessed 15 Feb. 2023].
#Renotte, N. (2022). Build a Deep Audio Classifier with Python and Tensorflow. [online] www.youtube.com. Available at: https://youtu.be/ZLIPkmmDJAc [Accessed 20 Feb. 2023].
#Stack Overflow. (n.d.). python - How to plot the accuracy and and loss from this Keras CNN model? [online] Available at: https://stackoverflow.com/questions/66785014/how-to-plot-the-accuracy-and-and-loss-from-this-keras-cnn-model [Accessed 12 Mar. 2023].

import os
import librosa
import numpy as np

#loopiing through the dataset path
for i in ['dataset\YAF_happy','dataset\YAF_sad','dataset\YAF_pleasant_surprised','dataset\YAF_neutral','dataset\YAF_fear','dataset\YAF_disgust','dataset\YAF_angry','dataset\OAF_Pleasant_surprise','dataset\OAF_neutral','dataset\OAF_Fear','dataset\OAF_disgust','dataset\OAF_angry','dataset\OAF_happy','dataset\OAF_sad']:
    audio_dir = i

    #loop through the files in the folder
    for file_name in os.listdir(audio_dir):
        if file_name.endswith('.wav'):
            #loading the audio file
            audio_path = os.path.join(audio_dir, file_name)
            audio, sr = librosa.load(audio_path, sr=44100)

            #processing the audio clip to 5 seconds
            duration = 5 # in seconds
            samples = sr * duration
            audio = librosa.resample(audio, orig_sr=sr, target_sr=22050)
            if len(audio) > samples:
                audio = audio[:samples]
            else:
                audio = np.pad(audio, (0, samples - len(audio)), mode='constant')

            #extract the features
            mfccs = librosa.feature.mfcc(y=audio, sr=22050, n_mfcc=13, n_fft=2048, hop_length=512)

            #save the mfccs
            mfcc_path = os.path.join(audio_dir, file_name.replace('.wav', '.npy'))
            np.save(mfcc_path, mfccs)
```

```python
import os
import tensorflow.keras as keras
import numpy as np
import tensorflow as tf 
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from keras.models import Sequential
from sklearn.preprocessing import LabelEncoder
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Reshape, LSTM
from matplotlib import pyplot as plt
mfccs = []
labels = []

for i in ['dataset\YAF_happy','dataset\YAF_sad','dataset\YAF_pleasant_surprised','dataset\YAF_neutral','dataset\YAF_fear','dataset\YAF_disgust','dataset\YAF_angry','dataset\OAF_Pleasant_surprise','dataset\OAF_neutral','dataset\OAF_Fear','dataset\OAF_disgust','dataset\OAF_angry','dataset\OAF_happy','dataset\OAF_sad']:
    audio_dir = i
    mfcc_dir =i

    #loading the files in the folder
    for file_name in os.listdir(mfcc_dir):
        if file_name.endswith('.npy'):
            #loading the mfccs
            mfcc_path = os.path.join(mfcc_dir, file_name)
            mfcc = np.load(mfcc_path)

            #append the mfccs to the list
            mfccs.append(mfcc)
            #files are names as 0 and 1 such that the label is 0 or 1
            labels.append(file_name.split('(')[0])  
#convert the mfccs and labels to a numpy array
mfccs = np.array(mfccs)
labels = np.array(labels)
#encoding the labels
lb=LabelEncoder()
labels=lb.fit_transform(labels)

#splitting to train and test sets
X_train, X_test, y_train, y_test = train_test_split(mfccs, labels, test_size=0.2, random_state=42,shuffle=True)


#reshaping the mfccs to the proper shape so its compatible with the model
n_frames = X_train.shape[1]
n_mfcc = X_train.shape[2]
n_channels = 1
X_train = np.expand_dims(X_train, axis=-1)
X_test = np.expand_dims(X_test, axis=-1)
X_train = X_train.reshape((X_train.shape[0], n_frames, n_mfcc, n_channels))
X_test = X_test.reshape((X_test.shape[0], n_frames, n_mfcc, n_channels))


#creating the training model
model = Sequential()

model.add(LSTM(units=128, input_shape=(13, 431), return_sequences=True))
model.add(LSTM(units=64, return_sequences=True))
model.add(LSTM(units=32, return_sequences=False))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


model.summary()

# training the created model
history = model.fit(X_train, y_train, validation_data=(X_test, y_test),batch_size=32, epochs=100)


        
```

```python
#loss_plot
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```

```python
#accuracy_plot
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='lower right')
plt.show()
```

```python
#saving the model
model.save('model3.h5')
```
